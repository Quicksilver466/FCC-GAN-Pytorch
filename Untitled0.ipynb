```python
%matplotlib inline
import torch
import torchvision
from torchvision import datasets, transforms
from torch import nn
from torch import optim 
import numpy as np
import matplotlib.pyplot as plt
import time
from tqdm import tqdm
from generator_fcc import Generator
from discriminator_fcc import Discriminator
```

```python
class fccgan():
    def __init__(self):
        self.modelG=Generator()
        self.modelD=Discriminator()
        self.train_set=0
        self.test_set=0
        self.loss_func=nn.BCELoss()
        self.optimizerG=optim.Adam(self.modelG.parameters(),lr=0.0001,weight_decay=0.000001)
        self.optimizerD=optim.Adam(self.modelD.parameters(),lr=0.0001,weight_decay=0.000001)       
    
    def dataprep(self):
        transform = transforms.Compose([transforms.ToTensor(),transforms.Normalize((0.5,), (0.5,)),])
        trainset=datasets.MNIST('/content/gdrive/My Drive/Colab Notebooks/GAN_FCC',download=False,train=True,transform=transform)
        testset=datasets.MNIST('/content/gdrive/My Drive/Colab Notebooks/GAN_FCC',download=False,train=False,transform=transform)

        trainloader=torch.utils.data.DataLoader(trainset,batch_size=32,shuffle=True)
        testloader=torch.utils.data.DataLoader(testset,batch_size=32,shuffle=True)
        return [trainloader,testloader]
    
    def test(self):
        inp=self.randgen(1)
        ###
        self.modelG=self.modelG.to('cpu')
        self.modelD=self.modelD.to('cpu')
        ###
        self.modelG=self.modelG.eval()
        self.modelD=self.modelD.eval()
        op=self.modelG(inp)        
        with torch.no_grad():
            output_img=np.array(op[0])
            output_img=output_img.reshape(28,28)
            plt.imshow(output_img,cmap='gray');
            plt.show()

    def randgen(self,batch_size):
        latent_vec=torch.randn(batch_size,100)
        return latent_vec

    def model_save(self):
        torch.save(self.modelG,'/content/gdrive/My Drive/Colab Notebooks/GAN_FCC/Saved_Models/gan_fcc_1_modelG.pth')
        torch.save(self.modelG.state_dict(),'/content/gdrive/My Drive/Colab Notebooks/GAN_FCC/Saved_Models/gan_fcc_1_modelG_state_dict.pth')
        torch.save(self.modelD,'/content/gdrive/My Drive/Colab Notebooks/GAN_FCC/Saved_Models/gan_fcc_1_modelD.pth')
        torch.save(self.modelD.state_dict(),'/content/gdrive/My Drive/Colab Notebooks/GAN_FCC/Saved_Models/gan_fcc_1_modelD_state_dict.pth')

    def load_model(self):
        self.modelG=torch.load('/content/gdrive/My Drive/Colab Notebooks/GAN_FCC/Saved_Models/gan_fcc_1_modelG.pth')
        self.modelG.load_state_dict(torch.load('/content/gdrive/My Drive/Colab Notebooks/GAN_FCC/Saved_Models/gan_fcc_1_modelG_state_dict.pth'))
        self.modelD=torch.load('/content/gdrive/My Drive/Colab Notebooks/GAN_FCC/Saved_Models/gan_fcc_1_modelD.pth')
        self.modelD.load_state_dict(torch.load('/content/gdrive/My Drive/Colab Notebooks/GAN_FCC/Saved_Models/gan_fcc_1_modelD_state_dict.pth'))

    def training_loop(self):
        self.modelG=self.modelG.train()
        self.modelD=self.modelD.train()
        dataset=self.dataprep()
        train=dataset[0]
        test=dataset[1]

        n_epochs=30
        lossG_list=[]
        lossD_list=[]
        flag=0
        ###
        self.modelG=self.modelG.to('cuda')
        self.modelD=self.modelD.to('cuda')
        ###
        for epoch in range(n_epochs):
            tic=time.time()
            ###
            self.modelG=self.modelG.to('cuda')
            self.modelD=self.modelD.to('cuda')
            ###
            for img,label in tqdm(train):
                #freezing generator and training discriminator
                self.modelG=self.modelG.eval()
                self.modelD=self.modelD.train()
                #training on real images
                ###
                img=img.to('cuda')
                ###
                real_labels=torch.full(size=(img.shape[0],1),fill_value=1,dtype=torch.float32)
                ###
                real_labels=real_labels.to('cuda')
                ###
                output_D_real=self.modelD(img)
                self.optimizerD.zero_grad()
                loss_D_real=self.loss_func(output_D_real,real_labels)
                ##if (flag==0):
                ##    print('The loss of Disrciminator for real images at the very beginning is: {}'.format(loss_D_real))
                ##loss_D_real.backward()
                ##self.optimizerD.step()
                #training on generated images
                latent_vec=self.randgen(32)
                ###
                latent_vec=latent_vec.to('cuda')
                ###
                output_G=self.modelG(latent_vec)
                fake_labels=torch.full(size=(output_G.shape[0],1),fill_value=0,dtype=torch.float32)
                ###
                fake_labels=fake_labels.to('cuda')
                ###
                output_D_fake=self.modelD(output_G)
                ##self.optimizerD.zero_grad()
                loss_D_fake=self.loss_func(output_D_fake,fake_labels)
                loss_D=(loss_D_fake+loss_D_real)/2
                if (flag==0):
                    print('The loss of Disrciminator at the very beginning is: {}'.format(loss_D_fake))
                loss_D.backward(retain_graph=True)
                self.optimizerD.step()
                #freezing discriminator and training generator
                self.modelD=self.modelD.eval()
                self.modelG=self.modelG.train()
                #giving discriminator generated images
                latent_vec=self.randgen(32)
                ###
                latent_vec=latent_vec.to('cuda')
                ###
                output_G=self.modelG(latent_vec)
                fake_real_labels=torch.full(size=(output_G.shape[0],1),fill_value=1,dtype=torch.float32)
                ###
                fake_real_labels=fake_real_labels.to('cuda')
                ###
                output_D_fake=self.modelD(output_G)
                self.optimizerG.zero_grad()
                loss_G=self.loss_func(output_D_fake,fake_real_labels)
                if (flag==0):
                    print('The loss of Generator at the very beginning is: {}'.format(loss_G))
                loss_G.backward()
                self.optimizerG.step()
                flag+=1
            toc=time.time()
            print('Loss of Discriminator at the end of epoch {} is: {}'.format((epoch+1),loss_D))
            print('Loss of Generator at the end of epoch {} is: {}'.format((epoch+1),loss_G))
            print('Time taken by this epoch is: {}'.format(((toc-tic)*1000)))
            print('The output by generator after the epoch {} is: '.format(epoch+1))
            for i in range(10):
                self.test()
```

```python
def weights_init(m):
    classname = m.__class__.__name__
    if classname.find('Conv') != -1:
        nn.init.normal_(m.weight.data, 0.0, 0.02)
    elif classname.find('BatchNorm') != -1:
        nn.init.normal_(m.weight.data, 1.0, 0.02)
        nn.init.constant_(m.bias.data, 0)
```

```python
gan=fccgan()
```

```python
weights_init(gan.modelG)
weights_init(gan.modelD)
```

```python
gan.training_loop()
```

```python
gan.test()
```
